```python
import psycopg2
import pandas as pd
```

```python
employee_df = pd.read_csv('employees1.csv')
employee_df
```

```python
employee_df.fillna(0, inplace=True)
employee_df
```

```python
department_df = pd.read_csv('department1.csv')
department_df
```

```python
def create_database():
    conn = psycopg2.connect("host = 127.0.0.1 dbname= postgres user = postgres password = jojo")
    conn.set_session(autocommit = True)
    cur = conn.cursor()
    
    #create sparkify database with utfb encoding
    cur.execute("drop database if exists accounts5")
    cur.execute("create database accounts5")
    
    #close connection to default db
    conn.close()
    
    #coonnect to sparkify db
    conn = psycopg2.connect("host = 127.0.0.1 dbname = accounts5 user = postgres password = jojo")
    cur = conn.cursor()
    
    return cur, conn
```

```python
cur, conn = create_database()
```

```python
employee_table_create = ("""create table if not exists employee (
EMPNO int Primary Key,
ENAME varchar,
JOB varchar,
HIREDATE varchar,
SAL int,
COMM int,
STORE_ID int)""")
```

```python
cur.execute(employee_table_create)
conn.commit()
```

```python
department_table_create = ("""create table if not exists department(
DEPTNO int,
DNAME varchar,
LOC varchar)""")
cur.execute(department_table_create)
conn.commit()
```

```python
employee_table_insert = ("""insert into employee(
EMPNO,
ENAME,
JOB,
HIREDATE,
SAL,
COMM,
STORE_ID)
values (%s,%s,%s,%s,%s,%s,%s)
""")
```

```python
for i, row in employee_df.iterrows():
    cur.execute(employee_table_insert,list(row))
```

```python
conn.commit() 
```
